```python
import json

with open("csv/bio-sig-meta.txt") as src:
    meta = json.load(src)

min = 0
max = 0

for m in meta:
    try:
            minC = m["bands"][0]["histogram"]["min"]
            maxC = m["bands"][0]["histogram"]["max"]
            if minC < min: min = minC
            if maxC > max: max = maxC
    except KeyError:
            pass

print(min)
print(max)
```

```python
histo = dict()
step = (max - min)/ 256

for i in range(0,256):
        bin = i * step - ( - min)
        histo[bin] = 0
        
def add_to_step(src_bin, value):
    
    for i in range(0,256):
        bin = i * step - ( - min)
        if bin <= src_bin < bin + step:
            histo[bin] += value


def process_histo(histogram):
    histo_step = (histogram["max"]-histogram["min"])/256
    
    for i in range(0,256):
        bin = i * histo_step - ( - histogram["min"])
        add_to_step(bin, histogram["buckets"][i])


for m in meta:
    try:
            process_histo(m["bands"][0]["histogram"])
    except KeyError:
            pass        

print(histo)
```
